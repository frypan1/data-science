```python
data['Age']= np.where(data['Age'] == '0-17','child',np.where(data['Age'] == '18-25','young',np.where(data['Age'] == '26-35','middle',np.where(data['Age'] == '36-45','senior','old'))))
data
```

```python
gp = data.groupby(['Gender','Age']).agg({'Product_Category_2':pd.Series.mode , 'Product_Category_3':pd.Series.mode})
gp
```

```python
for i in ['F','M']:
  if i == 'F':
    for j,v in zip(['child','young','middle','senior','old'],[4,14,14,14,14]):
      data.loc[(data['Product_Category_2'].isna()) & (data['Gender']==i)&(data['Age'] == j),'Product_Category_2'] = float(v)
  else :
    for j,v in zip(['child','young','middle','senior','old'],[2,8,8,8,8]):
      data.loc[(data['Product_Category_2'].isna()) & (data['Gender']==i)&(data['Age'] == j),'Product_Category_2'] = float(v)
```
